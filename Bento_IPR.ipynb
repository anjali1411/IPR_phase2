```python
%load_ext autoreload
%autoreload 2
import bento as bt
import spatialdata as sd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import zipfile
import os

# Path to the ZIP file in your Google Drive
zip_file_path = '/content/drive/MyDrive/data.zip'

# Set the unzip directory to the same folder as the ZIP file
unzip_dir = os.path.dirname(zip_file_path)

# Unzip the file in the same directory
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_dir)

print(f"Data unzipped to: {unzip_dir}")
```

```python
import os

# Check the contents of the directory to confirm the .zarr file is present
os.listdir(unzip_dir)
```

```python
import spatialdata as sd

# Update the path to the correct .zarr file
sdata_path = os.path.join(unzip_dir, 'merfish_u2os.zarr')

# Read the Zarr file
sdata = sd.read_zarr(sdata_path)
sdata
```

```python
sdata = sd.bounding_box_query(
    sdata,
    axes=["x", "y"],
    min_coordinate=[0, 0],
    max_coordinate=[6000, 6000],
    target_coordinate_system="global",
)

sdata = bt.io.prep(sdata)

# Visualize data
plt.figure(figsize=(10, 10))
bt.pl.density(sdata)

# Filter genes
sdata = bt.ut.filter_by_gene(sdata, min_count=10)

# Spatial summary statistics
bt.tl.shape_stats(sdata)
bt.pl.shape_stats(sdata)

# RNAflux analysis
res = 0.1
bt.tl.flux(sdata, method="radius", res=res, recompute=True)

# Visualize RNAflux results
fig, ax = plt.subplots(figsize=(12, 12))
bt.pl.flux(sdata, res=res, ax=ax)
```

```python
bt.tl.fluxmap(sdata, res=res, min_count=100, plot_error=True)

fig, ax = plt.subplots(figsize=(12, 12))
bt.pl.fluxmap(sdata, palette=bt.colors.bento6, ax=ax)
```

```python
fluxmap_names = [s for s in sdata.shapes.keys() if s.startswith("fluxmap")]
bt.tl.comp(
    sdata,
    points_key="transcripts",
    shape_names=fluxmap_names,
)
bt.pl.comp(sdata, annotate=5)
```
